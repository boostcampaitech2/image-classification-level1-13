### making data class

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/opt/ml/input/data/train/train.csv") # 위치이동 후 실행
```

```python
data.head()
```

```python
# 분류
# gender - male[0], female[1]
# age - <30[0], 30<= and <60[1], 60<=[2]
# mask - mask[0], incorrect[1], nomask[2]
```

```python
# 학습데이터 경로만 가져오기
# path는 폴더이름임
# 내부파일들이 구성되어져 있음
train_folder_paths = data['path']
```

```python
train_folder_paths.head(10)
```

```python
import os
import glob
from string import digits
```

```python
total = list()

for path in train_folder_paths.tolist():
    split_text = path.split('_')
    id = split_text[0]
    gender = split_text[1]
    age = int(split_text[3])
    for i in glob.glob("/opt/ml/input/data/train/images/{}/*".format(path)):
        # id, gender, age, mask, path
        table = str.maketrans('', '', digits)
        mask = os.path.basename(i).split('.')[0].translate(table) # 숫자제거 #글자수로 자를시 jpg와 jpeg가 둘다 있어서 문제
        total.append([id, gender, age, mask, i])
```

```python
header = ['id', 'gender', 'age', 'mask', 'path']
```

```python
df = pd.DataFrame(data=total, columns=header)
```

```python
df.head(5)
```

```python
# gender - male[0], female[1]
df['gender_class'] = df['gender'].map({'male': 0, 'female': 1})
```

```python
# age - <30[0], 30<= and <60[1], 60<=[2]
df['age_class'] = df['age']
df.loc[df.age < 30, "age_class"] = 0
df.loc[df.age >= 30, "age_class"] = 1
df.loc[df.age >= 60, "age_class"] = 2
```

```python
# mask - mask[0], incorrect[1], nomask[2]
df['mask_class'] = df['mask'].map({'mask' : 0, 'incorrect_mask' : 1, 'normal' : 2})
```

```python
df.dtypes
```

```python
# gender_class 점검
print(df['gender_class'].count(), df.loc[df['gender_class']==0, 'gender_class'].count(), df.loc[df['gender_class']==1, 'gender_class'].count())
print(df.loc[df['gender_class']==0, 'gender_class'].count() + df.loc[df['gender_class']==1, 'gender_class'].count())
print(df['gender_class'].unique())
```

```python
# age_class 점검
print(df['age_class'].count(), df.loc[df['age_class']==0, 'age_class'].count(), df.loc[df['age_class']==1, 'age_class'].count(), df.loc[df['age_class']==2, 'age_class'].count())
print(df.loc[df['age_class']==0, 'age_class'].count() + df.loc[df['age_class']==1, 'age_class'].count() + df.loc[df['age_class']==2, 'age_class'].count())
print(df['age_class'].unique())
```

```python
print(df['mask_class'].count(), df.loc[df['mask_class']==0, 'mask_class'].count(), df.loc[df['mask_class']==1, 'mask_class'].count(), df.loc[df['mask_class']==2, 'mask_class'].count())
print(df.loc[df['mask_class']==0, 'mask_class'].count() + df.loc[df['mask_class']==1, 'mask_class'].count() + df.loc[df['mask_class']==2, 'mask_class'].count())
print(df['mask_class'].unique())
```

```python
df.head(10)
```

```python
df['total_class'] = [str(x) + str(y) + str(z) for x, y, z in zip(df['mask_class'], df['gender_class'], df['age_class'])]
```

```python
df.head(10)
```

```python
mask_ = [0, 1, 2]
age_ = [0, 1, 2]
gender_ = [0, 1]
key = [str(i)+str(j)+str(k) for i in mask_ for j in gender_ for k in age_]
value = list(range(18))
print(value)
total_class = {key[i]:value[i] for i in value}
print(total_class)
```

```python
df['total_class'].unique()
```

```python
df.head(10)
```

```python
df['total_class_18'] = df['total_class'].map(total_class)
```

```python
df.drop(columns=['total_class'], inplace=True)
```

```python
df.head(10)
```

```python
train_num = df['mask_class'].count()//10 * 9
valid_num = df['mask_class'].count() - train_num
print(train_num, valid_num)
```

```python
train_csv = df.iloc[:train_num,:]
valid_csv = df.iloc[train_num:,:]
```

```python
valid_csv
```

```python
train_csv.to_csv("/opt/ml/input/data/train/train_class.csv", mode='w', index=False)
valid_csv.to_csv("/opt/ml/input/data/train/valid_class.csv", mode='w', index=False)
df.to_csv("/opt/ml/input/data/train/total_class.csv", mode='w', index=False)
```

### EDA

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# 데이터 그려보기, 이미 폴더상에서 확인을 했지만 python으로 그려보기
fig = plt.figure(figsize=[20, 140])

# 1개의 아이디에 대해 7개 이미지 그리기 ( mask 5, no mask 1, incorrect 1 )
rows = 1
cols = 7

img_path = df[df["id"] == "000001"]["path"].tolist()

for i in range(rows * cols):
    ax = fig.add_subplot(rows, cols, i+1)
    img = mpimg.imread(img_path[i])
    ax.imshow(img)
    plt.title("{}".format(os.path.basename(img_path[i]).split('.')[0]))

plt.show()

# 중심이 되는 얼굴과 마스크가 중앙에 위치함
```

```python
count_by_mask = df.groupby('mask_class').mask.count()
print(count_by_mask)

x_ = ['Mask', 'Incorrect', 'NoMask']
plt.bar(x_, count_by_mask, width = 0.5)
plt.title('Mask Count')
plt.xlabel('Mask')
plt.ylabel('Count')
plt.show()

# Mask가 압도적으로 많음
```

```python
count_by_gender = df.groupby('gender_class').gender.count()
print(count_by_gender)

x_ = ['Male', 'Female']
plt.bar(x_, count_by_gender, width = 0.3)
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# 여성이 더 많음
```

```python
count_by_age = df.groupby('age_class').age.count()
print(count_by_age)

x_ = ['Under 30', 'Bet 30 and 60', 'Over 60']
plt.bar(x_, count_by_age, width = 0.5)
plt.title('Age Count')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 60세 이상이 너무 적음
```

```python
count_by_age = df.groupby('age').age.count()
print(count_by_age)

x_ = df['age'].unique()
x_.sort()
plt.barh(x_, count_by_age)
plt.title('Age Count')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 나이 상세 분석 # 20세 전후, 50대가 많음
```

```python
count_by_total = df.groupby('total_class_18').total_class_18.count()
print(count_by_total)

x_ = df['total_class_18'].unique()
x_.sort()
plt.bar(x_, count_by_total)
plt.title('Class Count')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(x_)
plt.show()

# 4번 클래스가 제일 많음 -> 마스크 쓴 여자 30세~60세
```
